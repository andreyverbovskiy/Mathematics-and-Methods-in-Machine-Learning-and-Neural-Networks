```python
#3
x = np.array([np.arange(1,11,1)]).reshape(10,1)
y = np.array([[14],[12],[34],[24],[43],[38],[41],[57],[55],[50]])

X = np.array([[1, 1 ,1**2],[1,2,2**2],[1,3,3**2],[1,4,4**2],[1,5,5**2],[1,6,6**2],[1,7,7**2],[1,8,8**2],[1,9,9**2],[1,10,10**2]])

Xtr = X.transpose()

M = np.linalg.inv(np.dot(X.T, X)).dot(X.T).dot(y)

z = X.dot(M)
plt.plot(x, z)
plt.plot(x, y, 'o')
plt.show()
```
